```python
files=glob.glob('dataset3/*.csv')
print("Кількість файлів:")
print(len(files))
```

```python
dataf=[]
for f in files:
    try:
        df=pd.read_csv(f)
        dataf.append(df)
    except Exceptation as e:
        print(e)
```

```python
final_df = pd.concat(dataf, ignore_index=True)
final_df = final_df.sort_values(by='timestamp').reset_index(drop=True)
final_df['minute'] = final_df.index + 1
final_df.to_csv('final_data3_3.csv', index=False)
```
